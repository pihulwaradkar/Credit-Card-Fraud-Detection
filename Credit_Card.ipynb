```python
from tkinter import *
import joblib
import os
from tkinter import messagebox

def show_entry_fields():
    try:
        # Retrieve and validate values from entries
        values = []
        for e in entries:
            value_str = e.get()
            if not value_str:
                raise ValueError("Empty field detected")
            values.append(float(value_str))
    except ValueError as ve:
        messagebox.showerror("Input Error", f"Invalid input: {ve}")
        return

    # Print the current working directory for debugging
    current_dir = os.getcwd()
    print(f"Current working directory: {current_dir}")
    
    # Print the list of files in the current directory
    files = os.listdir(current_dir)
    print(f"Files in the directory: {files}")
    
    # Check if the model file exists
    model_path = 'credit_card_model'
    if not os.path.isfile(model_path):
        error_message = f"The model file '{model_path}' was not found in the directory '{current_dir}'."
        print(error_message)
        messagebox.showerror("File Not Found", error_message)
        return

    # Load the model and make predictions
    try:
        model = joblib.load(model_path)
        y_pred = model.predict([values])
        
        result = "Normal Transaction" if y_pred == 0 else "Fraudulent Transaction"
        
        print("######################################")
        print("Credit Card Fraud Detection System:", result)
        print("######################################")
        
        Label(master, text="Final Prediction from the model - credit card fraud detection").grid(row=31)
        Label(master, text=result).grid(row=32)
    except Exception as e:
        error_message = f"An error occurred while loading the model: {e}"
        print(error_message)
        messagebox.showerror("Error", error_message)

master = Tk()
master.title("Credit Card Fraud Detection System")

Label(master, text="Credit Card Fraud Detection System", bg="black", fg="white", width=30).grid(row=0, columnspan=2)

labels = [Label(master, text=f"Enter value of V{i+1}") for i in range(29)]
entries = [Entry(master) for _ in range(29)]

for i, (label, entry) in enumerate(zip(labels, entries)):
    label.grid(row=i+1, column=0)
    entry.grid(row=i+1, column=1)

Button(master, text='Predict', command=show_entry_fields).grid(row=30, column=1, sticky=W, pady=4)

mainloop()
```
